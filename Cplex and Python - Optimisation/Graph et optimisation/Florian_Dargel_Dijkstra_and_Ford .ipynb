```python
import numpy as np
infini = np.inf  
A = np.array([[infini,10,3,infini,infini],
             [infini,infini,1,2,infini],
            [infini,4,infini,infini,2],
            [infini,infini,infini,infini,7],
            [infini,infini,infini,9,infini]])
A 
```

```python
def algodijkstra(Matrice, depart):

#Etape 1 : Initialisation

    infini = np.inf 
    distances = [infini for i in range(len(Matrice))] #On met tous les distances à l'infini 
    distances[depart] = 0 #la distance du noeud de départ vaut 0 
    visited = [False for i in range(len(Matrice))] #Liste qui nous dit si le noeud a déjà été visité (Composé de False)
    n_visited = [] #les noeuds visités 

#Etape 2 : Trouver la distance minimale. 

    while 1: # Tant qu'il reste des noeuds à visiter (équalivalent à while True:)

        #On cherche le nœud avec la distance la plus courte du nœud de départ
        Pluscourt_distance = infini 
        Pluscourt_ = -1
        for i in range(len(Matrice)): 
            if distances[i] < Pluscourt_distance and not visited[i]:
                Pluscourt_distance = distances[i]
                Pluscourt_ = i
        n_visited.append(Pluscourt_) #On ajoute le sommet visité dans la liste        
        print("Noeuds visités " + str(Pluscourt_) + " avec la distance minimale " + str(Pluscourt_distance))
        

        
    
        if Pluscourt_ == -1:  # S'Il n'y a plus de noeud à visiter
            return "Valeur des sommets: " + str(distances) + " chemin " + str(n_visited[:-1]) 

          
#Etape 3 : Pour les noeuds voisins non visités / le nouveau chemin le plus court 

        for i in range(len(Matrice)):
            # si le chemin sur ce bord est plus court, on choisit ce chemin comme étant le nouveau chemin le plus court
            if Matrice[Pluscourt_][i] != 0 and distances[i] > distances[Pluscourt_] + Matrice[Pluscourt_][i]: 
                distances[i] = distances[Pluscourt_] +  Matrice[Pluscourt_][i]  
                print("Mise à jour distance du noeud " + str(i) + " à " + str(distances[i]))    
        visited[Pluscourt_] = True # on dit qu'on a fini avec ce nœud. 
        print("Noeuds visités: " + str(visited))
        print("Valeur des sommets: " + str(distances))
```

## Algorithme [Bellman-Ford] 

Dijkstra ne fonctionne pas avec les graphes contenants des arcs avec des valeurs négatives. Pour cela on peut appliquer la méthode de Bellman-Ford

```python
im3 = Image.open('C:/Users/na_to/OneDrive/Bureau/Insa/Mapromo/Graph_optimisation/ford.PNG')
im3 
```

```python
arcs_bis = ["ab","ac","bd","be","bf","cb","cf","ec","ed","fe"]
poids_bis = [7,8,4,1,2,5,2,-2,2,3]
sommets_bis = ["a","b","c","d","e","f"]
depart_bis = "a"
```

```python
import numpy as np

## Algo de Ford
def algo_bellman_ford(depart,arcs,sommet,poid,N=100000):
    
    #1ère étape : Initialisation 
    dist = [N for i in range(len(sommet))]# distance qui prend une très grande valeur : 100000 sur tous les sommets
    dist[sommet.index(depart)]=0 #le point de départ à pour valeur dist=0
    dist=np.array(dist) #On transforme la liste sous forme de tableau   
    predecesseurs=[depart if arcs[i][0]==depart else None for i in range(len(sommet))]# liste des précesseurs
  
   

    #2ème étape : Le coeur de l'algorithme
    boolean=True #condition boolean pour dire de s'arreter lorsque l'on obtient la même distance sur 2 itération.
    j=0
    while(boolean and j<=len(sommet)-1 ):
        j=j+1
        d_pred=dist
        for i in range(len(arcs)):
            a=int(list(np.where([ sommet[j]==arcs[i][0] for j in range(len(sommet))]))[0])
            b=int(list(np.where([ sommet[j]==arcs[i][1] for j in range(len(sommet))]))[0])
            if dist[b]-dist[a]>poid[i]:
                dist[b]=dist[a]+poid[i]
                predecesseurs[b]=sommet[a]
                

        boolean=(sum(d_pred==dist)!=len(dist))
        
    return({"les prédecesseurs sont":predecesseurs,"les potentiels sont":dist.tolist(),})

    
```

```python
algo_bellman_ford(depart_bis,arcs_bis,sommets_bis,poids_bis,N=100000)
```
